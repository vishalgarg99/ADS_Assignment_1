```python
# # Remove duplicate rows
df_cleaned = df.drop_duplicates()

# # Remove unnecessary columns
columns_to_drop = ['sub_category']
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
# Remove currency symbols and convert price columns to numeric
df_cleaned['discount_price'] = df['discount_price'].replace('[\₹,]', '', regex=True).astype(float)
df_cleaned['actual_price'] = df['actual_price'].replace('[\₹,]', '', regex=True).astype(float)
```

```python
# Convert numeric columns to the appropriate data type
df_cleaned['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')
df_cleaned['no_of_ratings'] = pd.to_numeric(df['no_of_ratings'], errors='coerce')
```

```python
# Descriptive statistics for numeric columns
numeric_columns = ['ratings', 'no_of_ratings']
numeric_data = df_cleaned[numeric_columns].replace([np.inf, -np.inf], np.nan).apply(pd.to_numeric, errors='coerce')

# Plot histograms for each numeric column
def hist():
    plt.figure(figsize=(15, 8))
    for i, column in enumerate(numeric_columns, 1):
        plt.subplot(2, 2, i)
        sns.histplot(numeric_data[column].dropna(), kde=True, bins=30, color='skyblue', edgecolor='black')
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
    return()
```

```python
# Group the DataFrame by 'main_category', count occurrences, and sort in descending order
popular_category = df.groupby('main_category',as_index=False).size().sort_values('size',ascending=False)
popular_category
```

```python
# Extract the first 10 values from 'main_category' column in df_cleaned
manufact = df_cleaned.iloc[:10]['main_category'].to_list() 

# Select rows from df where 'main_category' is in the manufact list
top = df[df['main_category'].isin(manufact)]
```

```python
def scatterplot():
    import plotly.express as px
#scatterplot to see if there is any relationship (with the trendline)
    fig = px.scatter(top, x="ratings", y="no_of_ratings", 
                 trendline="ols")

    fig.update_yaxes(range=[0, 2000])
    fig.update_layout(title = "Relationship between the number of reviews and Ratings",
                 yaxis_title = "Number of reviews",
                 xaxis_title = "Ratings")

    fig.show()
    return
```

```python
scatterplot()
```

```python
def violin():
#violion chart to see the distribution of the ratings
    plt.figure(figsize=(8,6))
    sns.violinplot(data=top, x='ratings',bins=30)

    plt.title('Distribution of ratings of the top 10',fontdict={'fontsize':13})
    plt.show()
    return
```

```python
violin()
```
